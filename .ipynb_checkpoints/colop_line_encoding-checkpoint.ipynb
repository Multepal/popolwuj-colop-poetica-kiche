```python
from bs4 import BeautifulSoup
from bs4 import SoupStrainer
import re
```

```python
colop_pw = 'colop-pk.xml'
```

```python
with open(colop_pw, 'r') as colop:
    data = colop.read()
```

```python
body_tag = SoupStrainer('body')
```

```python
soup = BeautifulSoup(data, 'xml', parse_only=body_tag)
```

```python
with open ('colop_pk_text.txt', 'a') as colop_pk_text:
    colop_pk_text.write(soup.get_text())
```

```python
with open('colop_pk_text.txt') as colop_text:
    text = colop_text.readlines()
```

```python
for i in range(len(text)):
    text[i] = re.sub(r'(\s{16})','', text[i], count=1)
```

```python
test_line = text[23]
```

```python
re.match(r'(\s+)', test_line).end(0)
```

```python
for i in range(25):
    offset = re.match(r'(\s+)', text[i])
    if offset is not None:
        quantity = offset.end(0)
        text[i] = text[i].strip()
        text[i] = '<space quantity="' + str(quantity) + '" unit="chars" /><l>' + text[i] + "</l>"
    else:
        text[i] = '<l>' + text[i] + '</l>'
```
